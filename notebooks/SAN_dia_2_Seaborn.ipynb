```python
import numpy as np
import pandas as pd

# Algunas librerias para graficar
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

# Seteamos el estilo
sns.set(style="darkgrid")# Usar ggplot?
```

# Introducción

Usaremos un dataset de jueguete para ejemplificar estas funciones rapidamente: [Iris Dataset](https://en.wikipedia.org/wiki/Iris_flower_data_set). 

El conjunto de datos contiene 50 muestras de cada una de tres especies de Iris (*Iris setosa*, *Iris virginica* e *Iris versicolor*). Se midió cuatro rasgos de cada muestra (en centímetros): 
* Largo del sépalo
* ancho del sépalo
* Largo del pétalo
* ancho del pétalo  

```python
url = 'https://raw.githubusercontent.com/gastonbujia/curso-visualizacion-datos-SAN-2022/main/data/iris.csv'
iris = pd.read_csv(url)
iris.rename(columns={'Name':'species'},inplace=True)
iris.head()
```

```python
iris.shape
```

```python
iris.info()
```

```python
iris.describe(include='all')
```

```python
metrics = ['count', 'min', 'max', 'mean','std','skew']
iris.groupby(by='species').agg(metrics)
```

```python
print(iris['species'].nunique())
print(iris['species'].unique())
```

## Boxplot

```python
# Boxplot de distribucion de cada columna numérica
sns.boxplot(data=iris);
```

Vamos a hacer un poco de magia y vamos a juntar 

```python
sns.violinplot(x="species", y="sepal_length", data=iris, height=5);
```

## Scatterplot

```python
# Scatterplots cuando las variables son categoricas
sns.swarmplot(x="species", y="petal_length", data=iris)
```

```python
#
sns.scatterplot(x="sepal_length", y="sepal_width", hue="species", data=iris)
```

## Histplot

```python
# un ejemplo mas de hist
```

```python

feats = iris.drop(columns=['species'])
for col in feats.columns:
  fig, (ax1,ax2) = plt.subplots(1, 2, figsize=(10,5), sharex=False)
  sns.histplot(iris[col],kde=True,ax=ax1)
  sns.violinplot(data=iris,x='species',y=col,inner='box',ax=ax2)
```

## Pairplot

```python
sns.pairplot(iris, hue="species", height=3);
```

## Displot

```python
sns.displot(data=iris, x="petal_length", hue="species", kind="kde");
```

## Heatmap

```python
iris_wo_spec = iris.drop(columns=['species'])
sns.heatmap(iris_wo_spec.corr(), center=0)
```

## Y si quiero un grafico interactivo?? 
También tenemos de esos

```python
import plotly.express as px
fig = px.bar(x=[0, 1, 2, 3, 4], y=[0, 1, 4, 9, 16])
fig.show()
```

```python
fig = px.scatter(iris, x="sepal_length", y="sepal_width",color="species", hover_data=['petal_width'])
fig.show()
```
